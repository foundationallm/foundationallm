```python
def verify_container(db, container_name : str):
    try:
        container = db.get_container_client(container_name.strip())
        container.read()
    except Exception as e:
        print('ERROR: Container: ' + container_name + ' does not exist.')

def verify_cosmosdb():

    from azure.cosmos import CosmosClient
    from azure.core.credentials import AzureKeyCredential

    #check connectivity
    lease_container_name = config.get_value("FoundationaLLM:CosmosDB:ChangeFeedLeaseContainer", use_retry=False)
    containers = config.get_value("FoundationaLLM:CosmosDB:Containers", use_retry=False)
    database_name = config.get_value("FoundationaLLM:CosmosDB:Database", use_retry=False)
    monitored_containers = config.get_value("FoundationaLLM:CosmosDB:MonitoredContainers", use_retry=False)

    endpoint = config.get_value("FoundationaLLM:CosmosDB:Endpoint", use_retry=False)
    key = config.get_value("FoundationaLLM:CosmosDB:Key", use_retry=False)

    client = CosmosClient(url=endpoint, credential={'masterKey': key} )

    #check database exits
    try:
        db = client.get_database_client(database_name)
        db.read()

        #check containers exist
        for container_name in containers.split(','):
            verify_container(db, container_name)

        verify_container(db, lease_container_name)

        for container_name in monitored_containers.split(','):
            verify_container(db, container_name)

    except Exception as e:
        print('ERROR: Database: ' + database_name + ' does not exist.')
```

```python
def verify_openai(openai_type, endpoint, key, models):

    #check connectivity
    import openai

    if ( openai_type == "azure_ad"):
        from azure.identity import DefaultAzureCredential
        credential = DefaultAzureCredential()
        token = credential.get_token("https://cognitiveservices.azure.com/.default")   

        openai.api_type = "azure_ad"
        openai.api_key = token.token
        openai.api_base = endpoint
        openai.api_version = "2023-05-15"  # subject to change
    
    if (openai_type == "azure"):
        openai.api_type = "azure"
        openai.api_key = key
        openai.api_base = endpoint
        openai.api_version = "2023-05-15"  # subject to change

    if (openai_type == "openai"):
        openai.api_type = "openai"
        openai.api_key = key
        openai.api_base = "https://api.openai.com/v1/"
        openai.api_version = None

    prompt = 'hello world'
    messages = ['hello', 'world']
    input = 'hello world'

    for model in models:
        #ensure models exist
        name = model['name']
        type = model['type']
        sub_model = model['model']

        if ( type == 'completion'):

            if (openai_type == "azure" or openai_type == "azure_ad"):
                print(openai.Completion.create(prompt=prompt, max_tokens=5, deployment_id=name, model=sub_model))
            else:
                print(openai.Completion.create(model=sub_model, messages=[{"role": "user", "content": "Hello world"}]))
        
        if ( type == 'chat-completion'):

            if (openai_type == "azure" or openai_type == "azure_ad"):
                print(openai.ChatCompletion.create(prompt=prompt, max_tokens=5, deployment_id=name, model=sub_model))
            else:
                print(openai.ChatCompletion.create(model=sub_model, messages=[{"role": "user", "content": "Hello world"}]))

        if ( type == "embedding"):

            if (openai_type == "azure" or openai_type == "azure_ad"):
                print(openai.Embedding.create(input=input, max_tokens=5, deployment_id=name, model=sub_model))
            else:
                print(openai.Embedding.create(model=sub_model, messages=[{"role": "user", "content": "Hello world"}]))
```

```python
def verify_azure_entra_ui():
    
    from msal import PublicClientApplication
    from azure.identity import ClientSecretCredential, DefaultAzureCredential

    #verify the chat UI app
    client_id = config.get_value('FoundationaLLM:Chat:Entra:ClientId', use_retry=False)
    client_secret = config.get_value('FoundationaLLM:Chat:Entra:ClientSecret', use_retry=False)
    tenant_id = config.get_value('FoundationaLLM:Chat:Entra:TenantId', use_retry=False)
    scope = f'{client_id}/.default'
    
    #check application exists
    app = PublicClientApplication(client_id, authority=f"https://login.microsoftonline.com/{tenant_id}")

    accounts = app.get_accounts()

    if ( len(accounts) > 0):
        result = app.acquire_token_silent(scope, account=accounts[0])

    #validate the client secret
    credential = ClientSecretCredential(
       tenant_id=tenant_id,
       client_id=client_id,
       client_secret=client_secret,
   )
    
    token = credential.get_token(scope)

    print(token)

    return token.token
```
